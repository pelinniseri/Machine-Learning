```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_csv("homeprices.xls")

df
```

```python
%matplotlib inline
plt.xlabel('Area (sqr fit)')
plt.ylabel('US $')
plt.scatter(df[['area']],df.price, color='red',marker='+')
```

```python
reg= linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

```python
df2=pd.read_csv("areas.xls")
df2
```

```python
pre_prices=reg.predict(df2)
```

```python
df2['PredictedPrices']=pre_prices
```

```python
df2
```

```python
df2.to_csv('Predicted Prices',index=False)
```

```python
%matplotlib inline
plt.xlabel('Area (sqr fit)')
plt.ylabel('US $')
plt.scatter(df2[['area']],df2.PredictedPrices, color='red',marker='+')
plt.plot(df2.area,reg.predict(df2[['area']]), color='blue')
```
